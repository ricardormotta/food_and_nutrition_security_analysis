```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("./data/treated_data.xlsx")
data_dict = pd.read_csv("./data/data_dict.csv", sep=";", encoding ='latin1')
```

```python
cols = df.columns.to_list()
for i in range(len(cols)):
    cols[i] = cols[i].strip()
df.columns = cols
data_dict.loc[:,'data_cod'] = data_dict.loc[:,'data_cod'].str.strip()

for col in df.columns:
    df.loc[df.loc[:,col] == "-", col] = np.nan
    
df.replace(' ', np.nan, inplace=True)
df.replace('X', np.nan, inplace=True)
```

```python
def get_year_from_data(df_data_dict, data_cod):
    return df_data_dict.loc[df_data_dict.data_cod == data_cod].year.iloc[0]

def medida_tipo_1(
    s1: pd.Series, 
    s2: pd.Series, 
    mult: int):
    return (s1 / s2) * mult

def medida_tipo_2(
    s1: pd.Series, 
    s2: pd.Series,
    s3: pd.Series,
    mult: int
):
    return mult * (s1 / (s2 + s3) )

def medida_tipo_3(
    s1: pd.Series, 
    s2: pd.Series,
    s3: pd.Series,
    mult: int
):
    return mult * ((s1+s2) / s3)
```
