```python
def get_year_from_data(df_data_dict, data_cod):
    return df_data_dict.loc[df_data_dict.data_cod == data_cod].year.iloc[0]

def medida_tipo_1(
    s1: pd.Series, 
    s2: pd.Series, 
    mult: int):
    return (s1 / s2) * mult

def medida_tipo_2(
    s1: pd.Series, 
    s2: pd.Series,
    s3: pd.Series,
    mult: int
):
    return mult * (s1 / (s2 + s3) )

def medida_tipo_3(
    s1: pd.Series, 
    s2: pd.Series,
    s3: pd.Series,
    mult: int
):
    return mult * ((s1+s2) / s3)
```

```python
# Critério quartil Crescente:
# Acima quartil 75% (incluso: Ótimo
# Entre quartil 50% (incluso) e quartil 75%: Bom
# Entre quartil 25% (incluso) e quartil 50%: Regular
# Abaixo quartil 25%: Ruim
parametros_quartis_crescentes = [
    'M7',
    'M8',
    'M9',
    'M11',
    'M12',
    'M13',
    'M17',
    'M18',
    'M19',
    'M20',
    'M21',
    'M22',
    'M23',
    'M24',
    'M25',
    'M26',
    'M27',
    'M28',
    'M30',
    'M32',
    'M35',
    'M44',
    'M47',
    'M48',
    'M49',
    'M50',
    'M51',
]
quartis = ['min','25%', '50%', '75%','max']
notas = [0, 4, 7, 10]

for M in parametros_quartis_crescentes:
    for i in range(len(notas)):
        sup = df[M].describe().loc[quartis[i+1]]
        inf = df[M].describe().loc[quartis[i]]
        df.loc[
            (
                (df[M] < sup) &
                (df[M] >= inf)
            ), 
            ('P' + M[1:])
        ] = notas[i]
```

```python
# Critério quartil Derescente:
# Acima quartil 75% (incluso: Ruim
# Entre quartil 50% (incluso) e quartil 75%: Regular
# Entre quartil 25% (incluso) e quartil 50%: Bom
# Abaixo quartil 25%: Ótimo
parametros_quartis_decrescentes = [
    'M1',
    'M2',
    'M3',
    'M5', #M5 possui critérios adicionais
    'M6',
    'M10',
    'M16',
    'M29',
    'M31',
    'M32',
    'M36',
    'M37',
    'M38',
    'M39',
    'M40',
    'M41',
    'M42',
    'M43',
    'M45',
    'M46',
    'M52', #M52 possui critérios adicionais
    'M53', #M53 possui critérios adicionais
    'M54', #M54 possui critérios adicionais
]
quartis = ['min','25%', '50%', '75%','max']
notas = [10, 7, 4, 0]

for M in parametros_quartis_decrescentes:
    for i in range(len(notas)):
        sup = df[M].describe().loc[quartis[i+1]]
        inf = df[M].describe().loc[quartis[i]]
        df.loc[
            (
                (df[M] < sup) &
                (df[M] >= inf)
            ), 
            ('P' + M[1:])
        ] = notas[i]
    #Todos que foram iguais ao minimo são ótimos:
    minimal = df[M].describe().loc[quartis[0]]
    df.loc[
            df[M]==minimal, 
            ('P' + M[1:])
        ] = notas[0]
```

```python
# Critério booleano positivo:
# Se >=1, então ótimo. 
# Se não, ruim.
parametros_booleanos_positivos = [
    'M4',
    'M33',
    'M34',
    'M55',
    'M55',
    'M56',
    'M57',
    'M59',
    'M58', #M58 possui critérios adicionais
]
notas = [0,1]
for M in parametros_booleanos_positivos:
    df.loc[:, ('P' + M[1:])] = notas[0]
    df.loc[
            df[M]>=1, 
            ('P' + M[1:])
        ] = notas[1]
```
